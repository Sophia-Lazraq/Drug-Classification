<h2> Introduction </h2>

<p>Chemotherapy is one of the most used treatment against cancer. It uses chemical substances (<a href=https://en.wikipedia.org/wiki/List_of_chemotherapeutic_agents>chemotherapeutic agents</a>) which kill cells that divide too quickly. These chemical substances are often diluted in a particular solution and packaged in bags, diffusers, or syringes, before being administered. <a href=https://books.google.fr/books?id=EB00rD8AqaYC&pg=PA188&lpg=PA188&dq=wrong+chemotherapeutic&source=bl&ots=m7EfyG6A3J&sig=ZVa0hLqDPFe2iExV6FOREJztN8c&hl=en&sa=X&ved=0ahUKEwj_5ZinkarMAhUFnBoKHaJiAgAQ6AEIKDAC#v=onepage&q=wrong%20chemotherapeutic&f=false>Wrong medication</a>  (wrong chemotherapeutic agent or wrong concentration) can have major impacts for patients. To prevent wrong medication, some recent French regulations impose the verification of anti-cancer drugs before their administration. The goal is to check that they contain the good chemotherapeutic agent with the good dosage. 

<p><a href=https://en.wikipedia.org/wiki/Raman_spectroscopy>Raman spectroscopy</a> could be used to make this check, since, theoretically, i) each molecule has a specific spectral fingerprint by which the molecule can be identified; and ii) the Raman intensity increases with the concentration of the molecule. The main advantage of spectroscopy above other methods (for example, liquid chromatography) is that it is non-destructive and non-invasive (measures are made without opening the drug containers). However, this method is rarely used in hospital environment because of the complexity of the spectral signals to analyze. Automating the analysis of these spectral signals could significantly help. Eventually, a complete analytical system (from measuring Raman spectra to identifying the chemotherapeutic agent and its concentration) could be designed, which would be easy to use and would prevent wrong medication. 

<p>In this context, the goal of this project is to develop prediction models able to <b>identify and quantify chemotherapeutic agents from their Raman spectra</b>.  

<p>The Lip(Sys)² laboratory measured Raman spectra of 4 types of chemotherapeutic agents (called <i>molecule</i>) in 3 different packages (called <i>vial</i>), diluted in 9 different solutions (called <i>solute gammes</i>), and having different concentrations. A total of <b>360 spectra were measured for each agent</b>, except for one (348 spectra).<br>  

Part of these data are saved in the file <code>train.csv</code> as follows (<code>n_samples</code> being the number of samples): 
<ul>
    <li><b>molecule</b>: Type of chemotherapeutic agent. Six possible values: A for infliximab, B for bévacizumab, Q for ramucirumab, R for rituximab. Dimension: (<code>n_samples</code>,)</li>
    <li><b>vial</b>: Vial type. Three possible values: 1, 2, 3. Dimension: (<code>1</code>, <code>n_samples</code>)</li>
    <li><b>solute</b>: Solute group. Fourteen possible values: 1, 2, ..., 14. Dimension: (<code>1</code>, <code>n_samples</code>)</li>
    <li><b>concentration</b>: Concentration of the molecule. Dimension: (<code>n_samples</code>, <code>1</code>)</li>
    <li><b>spectra</b>: Intensity of Raman spectrum. Dimension: (<code>n_samples</code>, <code>1866</code>)</li>
</ul>

<p>To sum up, there are too objectives:

<ul>
    <li><b>classification</b>: predict which molecule it corresponds to given the spectrum.</li>
    <li><b>regression</b>: predict the concentration of a molecule. The prediction should not depend on the vial or the solute group. The error metric is the mean absolute relative error (mare): $$\frac{1}{n_{samples}}\sum_{i=1}^{n_{samples}}\left|\frac{y_i-\hat{y}_i}{y_i}\right|$$ with $y$ and $\hat{y}$ being the true and predicted concentration.</li>
</ul>

<h2> Introduction </h2>

<p> La chimiothérapie est l'un des traitements les plus utilisés contre le cancer. Il utilise des substances chimiques (<a href=https://en.wikipedia.org/wiki/List_of_chemotherapeutic_agents> agents chimiothérapeutiques </a>) qui tuent des cellules qui se divisent trop rapidement. Ces substances chimiques sont souvent diluées dans une solution particulière et conditionnées dans des sacs, des diffuseurs ou des seringues avant d'être administrées.<a href=https://books.google.fr/books?id=EB00rD8AqaYC&pg=PA188&lpg=PA188&dq=wrong+chemotherapeutic&source=bl&ots=m7EfyG6A3J&sig=ZVa0hLqDPFe2iExV6FOREJztN8c&hl=en&sa=X&ved=0ahUKEwj_5ZinkarMAhUFnBoKHaJiAgAQ6AEIKDAC#v=onepage&q=wrong%20chemotherapeutic&f=false>Le mauvais médicament </a> (mauvais agent chimiothérapeutique ou mauvaise concentration) peut avoir des conséquences majeures pour les patients. Pour prévenir de mauvais médicaments, certaines réglementations françaises récentes imposent la vérification des médicaments anticancéreux avant leur administration. Le but est de vérifier qu'ils contiennent le bon agent chimiothérapeutique avec le bon dosage.

<p> <a href=https://en.wikipedia.org/wiki/Raman_spectroscopy> La spectroscopie Raman </a> pourrait être utilisée pour effectuer cette vérification, puisque, théoriquement, 
- i) chaque molécule a une empreinte spectrale spécifique par laquelle La molécule peut être identifiée;  
- ii) Et l'intensité Raman augmente avec la concentration de la molécule. Le principal avantage de la spectroscopie par rapport à d'autres méthodes (par exemple, la chromatographie liquide) est qu'il est non destructeur et non invasif (les mesures sont faites sans ouvrir les récipients de médicament). Cependant, cette méthode est rarement utilisée en milieu hospitalier en raison de la complexité des signaux spectraux à analyser. L'automatisation de l'analyse de ces signaux spectraux pourrait considérablement aider. Finalement, un système analytique complet (de la mesure des spectres Raman à l'identification de l'agent chimiothérapique et de sa concentration) pourrait être conçu, ce qui serait facile à utiliser et empêcherait un mauvais médicament.

<p> Dans ce contexte, le but de ce projet est de développer des modèles de prédiction capables d'<b>identifier et quantifier les agents chimiothérapeutiques à partir de leurs spectres Raman </b>.

<p> Le laboratoire Lip (Sys) ² a mesuré les spectres Raman de 4 types d'agents chimiothérapeutiques (appelé <i> molécule </i>) en 3 emballages différents (appelés <i>flacon </i>), dilués dans 9 (Appelés <i> solute gammes </i>), et ayant des concentrations différentes. <ul>

    <li> <b> flacon </b>: Type de flacon. Trois valeurs possibles: 1, 2, 3. Dimension: (<code> 1 </code>, <code> n_samples </code>) </li>
    <li> <b> soluté </b>: Groupe de solute. Quatorze valeurs possibles: 1, 2, ..., 14. Dimension: (<code> 1 </code>, <code> n_samples </code>) </li>
    <li> <b> concentration </b>: Concentration de la molécule. Dimension: (<code> n_samples </code>, <code> 1 </code>) </li>

</ul>

<p> Pour résumer, il y a trop d'objectifs:

<ul>
    <li> <b> classement </b>: prédire à quelle molécule il correspond selon le spectre. </li>

La métrique d'erreur est l'erreur relative absolue moyenne (mare): $$\frac{1}{n_{samples}}\sum_{i=1}^{n_{samples}}\left|\frac{y_i-\hat{y}_i}{y_i}\right|$$ with $y$ and $\hat{y}$ étant la concentration vraie et prédite. </li>

# CHOIX DU CLASSIFICATEUR

```python
# RF

from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator
from sklearn.neural_network import MLPClassifier
from sklearn.decomposition import PCA, KernelPCA



class Classifier_RF(BaseEstimator):
    def __init__(self):
        self.n_components = 10
        self.n_estimators = 300
        self.clf = Pipeline([
            ('pca', PCA(n_components=self.n_components)),
            ('clf', RandomForestClassifier(n_estimators=self.n_estimators, random_state=42))
        ])

    def fit(self, X, y):
        self.clf.fit(X, y)

    def predict(self, X):
        return self.clf.predict(X)

    def predict_proba(self, X):
        return self.clf.predict_proba(X)
```

```python
#neural network

from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator
from sklearn.neural_network import MLPClassifier



class Classifier_NN(BaseEstimator):
    def __init__(self):
        self.n_components = 10
        self.n_estimators = 300
        self.clf = Pipeline([
            ('pca', PCA(n_components=self.n_components)),
            ('clf', MLPClassifier(alpha=1))
                    ])

    def fit(self, X, y):
        self.clf.fit(X, y)

    def predict(self, X):
        return self.clf.predict(X)

    def predict_proba(self, X):
        return self.clf.predict_proba(X)
```

```python
# GaussianNB()
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator
from sklearn.naive_bayes import GaussianNB



class Classifier_GNB(BaseEstimator):
    def __init__(self):
        self.n_components = 10
        self.n_estimators = 300
        self.clf = Pipeline([
            ('pca', PCA(n_components=self.n_components)),
            ('clf',GaussianNB())
        ])

    def fit(self, X, y):
        self.clf.fit(X, y)

    def predict(self, X):
        return self.clf.predict(X)

    def predict_proba(self, X):
        return self.clf.predict_proba(X)
```

```python
# AdaBoostClassifier

from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier



class Classifier_ABoost(BaseEstimator):
    def __init__(self):
        self.n_components = 10
        self.n_estimators = 300
        self.clf = Pipeline([
            ('pca', PCA(n_components=self.n_components)),
            ('clf', AdaBoostClassifier())
        ])

    def fit(self, X, y):
        self.clf.fit(X, y)

    def predict(self, X):
        return self.clf.predict(X)

    def predict_proba(self, X):
        return self.clf.predict_proba(X)
```

```python
#DecisionTreeClassifier

from sklearn.tree import DecisionTreeClassifier
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator




class Classifier_DT(BaseEstimator):
    def __init__(self):
        self.n_components = 10
        self.n_estimators = 300
        self.clf = Pipeline([
            ('pca', PCA(n_components=self.n_components)),
            ('clf', DecisionTreeClassifier(max_depth=5))
        ])

    def fit(self, X, y):
        self.clf.fit(X, y)

    def predict(self, X):
        return self.clf.predict(X)

    def predict_proba(self, X):
        return self.clf.predict_proba(X)
```

```python
### QuadraticDiscriminantAnalysis

from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator

class Classifier_QDA(BaseEstimator):
    def __init__(self):
        self.n_components = 10
        self.n_estimators = 300
        self.clf = Pipeline([
            ('pca', PCA(n_components=self.n_components)),
            ('clf', QuadraticDiscriminantAnalysis())
        ])

    def fit(self, X, y):
        self.clf.fit(X, y)

    def predict(self, X):
        return self.clf.predict(X)

    def predict_proba(self, X):
        return self.clf.predict_proba(X)
```

```python
#GradientBoostingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator




class Classifier_GBoost(BaseEstimator):
    def __init__(self):
        self.n_components = 10
        self.n_estimators = 300
        self.clf = Pipeline([
            ('pca', PCA(n_components=self.n_components)),
            ('clf', GradientBoostingClassifier(n_estimators=self.n_estimators, learning_rate=1.0, max_depth=1, random_state=42)
)
        ])

    def fit(self, X, y):
        self.clf.fit(X, y)

    def predict(self, X):
        return self.clf.predict(X)

    def predict_proba(self, X):
        return self.clf.predict_proba(X)
```

```python
#LASSO
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator
from sklearn import linear_model


class Classifier_LASSO(BaseEstimator):
    def __init__(self):
        self.n_components = 10
        self.n_estimators = 300
        self.clf = Pipeline([
            ('pca', PCA(n_components=self.n_components)),
            ('clf', linear_model.LogisticRegression(penalty='l1', dual=False, tol=0.0001,
                C=200.0, fit_intercept=True, intercept_scaling=1, class_weight=None, random_state=None
            , solver='liblinear', max_iter=100, multi_class='ovr', verbose=0, warm_start=False, n_jobs=1))
        ])

    def fit(self, X, y):
        self.clf.fit(X, y)

    def predict(self, X):
        return self.clf.predict(X)

    def predict_proba(self, X):
        return self.clf.predict_proba(X)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

labels = np.array(['A', 'B', 'Q', 'R'])

def train_test_model_clf(X_df, y_df, skf_is, FeatureExtractor, Classifier):
    train_is, test_is = skf_is
    X_train_df = X_df.iloc[train_is].copy()                                  
    y_train_df = y_df.iloc[train_is].copy()
    y_train_clf = y_train_df['molecule'].values
    X_test_df = X_df.iloc[test_is].copy()                                    
    y_test_df = y_df.iloc[test_is].copy() 
    y_test_clf = y_test_df['molecule'].values 
    # Feature extraction
    fe_clf = FeatureExtractor()
    fe_clf.fit(X_train_df, y_train_df)
    X_train_array_clf = fe_clf.transform(X_train_df)
    X_test_array_clf = fe_clf.transform(X_test_df)
    # Train
    clf = Classifier()
    clf.fit(X_train_array_clf, y_train_clf)
    # Test 
    y_proba_clf = clf.predict_proba(X_test_array_clf)                        
    y_pred_clf = labels[np.argmax(y_proba_clf, axis=1)]                      
    error = 1 - accuracy_score(y_test_clf, y_pred_clf)                       
    print('error = %s' % error)                                                                            
    print('classification report:\n %s' % classification_report(y_test_clf, y_pred_clf))
    #print('confusion matrix:\n %s' % confusion_matrix(y_test_clf, y_pred_clf))


skf = ShuffleSplit(n_splits=2, test_size=0.2, random_state=57)  
skf_is = list(skf.split(X_df))[0]

print("METHODE RF")
train_test_model_clf(X_df, y_df, skf_is, FeatureExtractorClf, Classifier_RF)
print("************************************************************************************************************")
print("METHODE NN")
train_test_model_clf(X_df, y_df, skf_is, FeatureExtractorClf, Classifier_NN)
print("************************************************************************************************************")
print("METHODE Classifier_GNB")
train_test_model_clf(X_df, y_df, skf_is, FeatureExtractorClf, Classifier_GNB)
print("************************************************************************************************************")
print("METHODE Classifier_ABoost")
train_test_model_clf(X_df, y_df, skf_is, FeatureExtractorClf, Classifier_ABoost)
print("************************************************************************************************************")
print("METHODE Classifier_DT")
train_test_model_clf(X_df, y_df, skf_is, FeatureExtractorClf, Classifier_DT)
print("************************************************************************************************************")
print("METHODE Classifier_QDA")
train_test_model_clf(X_df, y_df, skf_is, FeatureExtractorClf, Classifier_QDA)
print("************************************************************************************************************")
print("METHODE Classifier_GBoost")
train_test_model_clf(X_df, y_df, skf_is, FeatureExtractorClf, Classifier_GBoost)
print("************************************************************************************************************")
print("METHODE Classifier_LASSO")
train_test_model_clf(X_df, y_df, skf_is, FeatureExtractorClf, Classifier_LASSO)
```

```python
# RF avec PCA

from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator
from sklearn.neural_network import MLPClassifier
from sklearn.decomposition import PCA, KernelPCA



class Classifier_RF_PCA(BaseEstimator):
    def __init__(self):
        self.n_components = 10
        self.n_estimators = 300
        self.clf = Pipeline([
            ('pca', PCA(n_components=self.n_components)),
            ('clf', RandomForestClassifier(n_estimators=self.n_estimators, random_state=42))
        ])

    def fit(self, X, y):
        self.clf.fit(X, y)

    def predict(self, X):
        return self.clf.predict(X)

    def predict_proba(self, X):
        return self.clf.predict_proba(X)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

labels = np.array(['A', 'B', 'Q', 'R'])

def train_test_model_clf(X_df, y_df, skf_is, FeatureExtractor, Classifier):
    train_is, test_is = skf_is
    X_train_df = X_df.iloc[train_is].copy()                                  
    y_train_df = y_df.iloc[train_is].copy()
    y_train_clf = y_train_df['molecule'].values
    X_test_df = X_df.iloc[test_is].copy()                                    
    y_test_df = y_df.iloc[test_is].copy() 
    y_test_clf = y_test_df['molecule'].values 
    # Feature extraction
    fe_clf = FeatureExtractor()
    fe_clf.fit(X_train_df, y_train_df)
    X_train_array_clf = fe_clf.transform(X_train_df)
    X_test_array_clf = fe_clf.transform(X_test_df)
    # Train
    clf = Classifier()
    clf.fit(X_train_array_clf, y_train_clf)
    # Test 
    y_proba_clf = clf.predict_proba(X_test_array_clf)                        
    y_pred_clf = labels[np.argmax(y_proba_clf, axis=1)]                      
    error = 1 - accuracy_score(y_test_clf, y_pred_clf)                       
    print('error = %s' % error)                                                                            
    print('classification report:\n %s' % classification_report(y_test_clf, y_pred_clf))
    #print('confusion matrix:\n %s' % confusion_matrix(y_test_clf, y_pred_clf))


skf = ShuffleSplit(n_splits=2, test_size=0.2, random_state=57)  
skf_is = list(skf.split(X_df))[0]
```

### Feature extractor for regression - Extracteur de fonctions pour la régression

Similarly to the feature extractor for classification, the feature extractor for regression should be implemented by a function <code>transform</code> and be part of the <code>FeatureExtractorReg</code> class, saved in the file <code>feature_extractor_reg.py</code>.  
<code>transform</code> receives the pandas dataframe <code>X_df</code> defined at the beginning of the notebook. The dataframe is augmented by four columns. At train time these columns contain the one-hot encoded molecule type; at test time, it receives the class posteriors. In this way you can use the classification information in the feature extractor of the regressor. The feature extractor should produce a numpy array representing the features extracted, which would be used for the regression.

In the example below, the features extracted are the molecule types and the standardized Raman spectra.

De même que l'extracteur de fonctions pour la classification, l'extracteur de fonctions pour la régression doit être implémenté par une fonction <code> transform </code> et faire partie de la classe <code> FeatureExtractorReg </code>, enregistré dans le fichier <code> feature_extractor_reg .py </code>.
<code> transform </code> reçoit la base de données pandas <code> X_df </code> définie au début du bloc-notes. La base de données est augmentée de quatre colonnes. A l'heure du train, ces colonnes contiennent le type de molécule encodée à chaud; Au moment du test, il reçoit la classe postérieure. De cette façon, vous pouvez utiliser les informations de classification dans l'extracteur de fonctions du régresseur. L'extracteur de caractéristiques devrait produire un tableau numpy représentant les fonctions extraites, qui serait utilisé pour la régression.

Dans l'exemple ci-dessous, les caractéristiques extraites sont les types de molécules et les spectres Raman normalisés.

```python
labels = np.array(['A', 'B', 'Q', 'R'])

class FeatureExtractorReg(object):
    def __init__(self):
        pass

    def fit(self, X_df, y):
        pass
    
    def transform(self, X_df):                                                   
        XX = np.array([np.array(dd) for dd in X_df['spectra']])                  
        XX -= np.median(XX, axis=1)[:, None]                                     
        XX /= np.sqrt(np.sum(XX ** 2, axis=1))[:, None]                          
        XX = np.concatenate([XX, X_df[labels].values], axis=1)                   
        return XX   
```

### Regression: predicting the concentration - Régression: prédiction de la concentration

The regressor follows a classical scikit-learn regressor template. It should be saved in the file <code>regressor.py</code>. In its simplest form it takes a scikit-learn pipeline and assignes it to <code>self.reg</code> in <code>__init__</code>.  

In the example below, one model is created for each type of molecule

As mentionned above, the error metric is the mean absolute relative error (mare): $$\frac{1}{n_{samples}}\sum_{k=1}^{n_{samples}}\left|\frac{y-\hat{y}}{y}\right|$$ with $y$ and $\hat{y}$ being the true and predicted concentration.

Le régresseur suit un modèle de régression classique scikit-learn. Il doit être enregistré dans le fichier <code>regressor.py</code>. Dans sa forme la plus simple, il prend un pipeline scikit-learn et l'affecte à <code>self.reg</code> in <code>_init_</code>

Dans l'exemple ci-dessous, un modèle est créé pour chaque type de molécule

Comme indiqué ci-dessus, la métrique d'erreur est l'erreur relative moyenne absolue (mare):
As mentionned above, the error metric is the mean absolute relative error (mare): $$\frac{1}{n_{samples}}\sum_{k=1}^{n_{samples}}\left|\frac{y-\hat{y}}{y}\right|$$ avec $y$ et $\hat{y}$ étant la concentration vraie et prédite.

```python
from sklearn.ensemble import GradientBoostingRegressor                           
from sklearn.decomposition import PCA                                            
from sklearn.pipeline import Pipeline                                            
from sklearn.base import BaseEstimator                                           
import numpy as np                                                               
                                                                                 
                                                                                 
class Regressor_PCA_GBoost(BaseEstimator):                                                  
    def __init__(self):                                                          
        self.n_components = 100                                                  
        self.n_estimators = 40                                                   
        self.learning_rate = 0.2                                                 
        self.list_molecule = ['A', 'B', 'Q', 'R']                                
        self.dict_reg = {}                                                       
        for mol in self.list_molecule:                                           
            self.dict_reg[mol] = Pipeline([                                      
                ('pca', PCA(n_components=self.n_components)),                    
                ('reg', GradientBoostingRegressor(                               
                    n_estimators=self.n_estimators,                              
                    learning_rate=self.learning_rate,                            
                    random_state=42))                                            
            ])                                                                   
                                                                                 
    def fit(self, X, y):                                                         
        for i, mol in enumerate(self.list_molecule):                             
            ind_mol = np.where(np.argmax(X[:, -4:], axis=1) == i)[0]             
            XX_mol = X[ind_mol]                                                  
            y_mol = y[ind_mol].astype(float)                                     
            self.dict_reg[mol].fit(XX_mol, np.log(y_mol))                        
                                                                                 
    def predict(self, X):                                                        
        y_pred = np.zeros(X.shape[0])                                            
        for i, mol in enumerate(self.list_molecule):                             
            ind_mol = np.where(np.argmax(X[:, -4:], axis=1) == i)[0]             
            XX_mol = X[ind_mol].astype(float)                                    
            y_pred[ind_mol] = np.exp(self.dict_reg[mol].predict(XX_mol))         
        return y_pred                                                                            
```
